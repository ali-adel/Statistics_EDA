# The Variance is a special Case of Covarinace (if you calculate the Covariance for the variable with itself you get the variance) So this Lead to the **Covariance Matrix**

# Covariance Matrix

```python
import numpy as np
x = [2 , 8 , 18 , 20 , 28 , 30 ]
y = [5 , 12 , 18 , 23 , 45 , 50 ]
```

```python
np.cov(x,y) # left diagonal is the var of each variable and the remaining is the Covariance (Symmetic Matrix)
```

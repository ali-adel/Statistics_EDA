# **Ex :** compare the kilometer that car take  before and after trying a type of oil for cars
* alpha = 0.05
* and i want to know if it affect positively

* H0 : that no difference Mu(diff) = 0
* H1 : there are a positive diff Mu(diff) > 0

```python
pre = [30, 31, 34, 40, 36, 35, 34, 30, 28, 29]
post = [30, 31, 32, 38, 32, 31, 32, 29, 28, 30]

ttest_rel(pre,post,alternative="greater")
```

**so (pvalue < alpha) so we reject H0 (there are significant diff)**

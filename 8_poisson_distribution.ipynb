```python
import numpy as np
from scipy.stats import poisson
import seaborn  as sns
import matplotlib.pyplot as plt
```

# theoritical

# Poisson_distribution

# it is concerned about average rate (lambda) or number of occurences per time unit  

image.png

**lambda  => average rate and it's equal the mean**

 **x => RV of number of occurences**

# Example

# if we have a restaurant and the average of customers that enter the restaurant is 2 customer / min

# so
  **lambda = mean  = 2**

  **x = number of hesitated () customers to restaurant / min**


```python
x = np.arange(0,10)
lambda_ = 2
prob =poisson.pmf(lambda_,x)
```

```python
prob[0]
```

```python
prob[1]
```

```python
prob[2]
```

```python
prob[3]
```

```python
bar = sns.barplot(x=x,y=prob)
```

# so as we notice the probalbility is increasing till reach lambda as it is equal to mean and decreasing after that and it is skewed for small lambda and approximately normal for larger values as the p(x) will be more close to each other  

# probability for lambda > 2

# practical ( with sample )

# here when we take more sample size, the mean for the sample data will be too close for theoritical mean of same data wih same lambda

```python
lambda2 = 2
data = np.random.poisson(lambda2,size=10)
data.mean()
```

```python
bar3 = sns.countplot(x=data)
```

```python
lambda2 = 2
data2 = np.random.poisson(lambda2,size=100)
data2.mean()
```

```python
bar4 = sns.countplot(x=data2)
```

```python
lambda2 = 2
data3 = np.random.poisson(lambda2,size=1000)
data3.mean()
```

```python
bar3 = sns.countplot(x=data3)
```

**here with sample size = 1000000 we notice that the mean is almost = 2**

```python
lambda2 = 2
data4 = np.random.poisson(lambda2,size=1000000)
data4.mean()
```

```python
bar4 = sns.countplot(x=data4)
```

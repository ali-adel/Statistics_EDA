# when i can't depend on the mean to know the center of data ?
**when we find outliers as it affect the mean and can't represent the data any more**

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/melb_data.csv')
```

```python
df.head()
```

```python
mu = df['Price'].mean()
mu
```

```python
df['Price'].max()
```

```python
plt.vlines(mu,colors='r',lw=2,ymin=0,ymax=900)
price_hist = sns.histplot(data=df,x='Price')
# we notice here that the mean is shifted to the right as there outliers take it to the right side (so we should think in another way to represent this data)
```

# Median

image.png

```python
median = df['Price'].median()
print("median is : " , median)
print("mean is : " , mu)
```

```python
plt.vlines(median,colors='b',lw=2,ymin=0,ymax=900)
plt.vlines(mu,colors='r',lw=2,ymin=0,ymax=900)
price_hist = sns.histplot(data=df,x='Price')
# we notice here that the median is more representative for this data as it very close from the actual center of data and doesn't affected by outliers
```

# Mode

it gvies The most frequent number in data and if all numbers is Repeated evenly so it doesn't have a mode

image.png

```python
mod = df['Price'].mode()
print("median is : " , median)
print("mean is : " , mu)
print("mode is : " , mod)
```

```python
plt.vlines(mu,colors='r',lw=2,ymin=0,ymax=900)
plt.vlines(median,colors='b',lw=2,ymin=0,ymax=900)
plt.vlines(mod,colors='black',lw=2,ymin=0,ymax=900)
price_hist = sns.histplot(data=df,x='Price')
# we notice here that 1100000 is the most frequent number here
```

```python
sample_1 = np.random.randint(1,7,size=10)
sample_1
```

```python
x_dash = sample_1.mean() # so we see here the big difference between this sample mean and the theoritical(population) mean as the sample is very small
x_dash
```

```python
sns.countplot(x=sample_1)
```
